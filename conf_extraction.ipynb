```python
import numpy as np
import liblzfse
import matplotlib.pyplot as plt
import sys
import cv2
path = './images/525.conf'
def load_conf(filepath):
        with open(filepath, "rb") as depth_fh:
            raw_bytes = depth_fh.read()
            decompressed_bytes = liblzfse.decompress(raw_bytes)
            depth_img = np.frombuffer(decompressed_bytes, dtype=np.uint8)
        if depth_img.shape[0] == 960 * 720:
            depth_img = depth_img.reshape((720, 960))  # For a FaceID camera 3D Video
        else:
            depth_img = depth_img.reshape((256, 192))  # For a LiDAR 3D Video
        return depth_img


if __name__ == '__main__':
    conf_original = load_conf(path)
    print(conf_img.shape)
    conf_original = cv2.resize(conf_img, (720, 960))

    plt.imshow(conf_original, 'jet')
    plt.colorbar()
    plt.show()
    print(np.unique(conf_original))
```
